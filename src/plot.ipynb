```python
import glob

lines = []

path = '../outputs/perplexity/wiki/output_perplexity_nuclear_nd_P28_2.txt'
path = '../outputs/perplexity/wiki/output_perplexity_cyber_2_P2_3.txt'
path = '../outputs/perplexity/wiki/output_perplexity_cyber_nd_P9_0.txt'
path = '../outputs/perplexity/wiki/output_perplexity_cyber_nd_P19_0.txt'

tokens = []
perplexities = []
with open(path, "r", encoding="utf-8") as file:
    b = False

    prompt = file.readline().split(": ")[1]

    for line in file:
        if "sequence of low perplexity" in line or "Generated" in line:
            break
        if len(line) > 4:
            lines.append(line)
        if b and ":" in line:
            print(line)
            tmp = line.split(": ")
            if len(tmp) > 3:
                token = tmp[1]
                perp_prob = tmp[2]
            else:
                token, perp_prob, _ = tmp 
            perp = perp_prob.split('(')[0]
            tokens.append(token.strip())
            perplexities.append(float(perp))

        if "perplexities:" in line:
            b = True
```
